```python
import json
from sklearn.metrics import confusion_matrix
```

```python
# JSONファイルを読み込む関数
def load_json(file_name):
    with open(file_name, 'r') as file:
        data = json.load(file)
    return data
```

```python
# データを抽出する関数
def extract_data(json_data, field):
    return [item[field] for item in json_data]
```

```python
# # ファイル名とフィールド名を定義
# file1_name = 'predata_confusionmatrix/A_train_converted.json'
# # file2_name = 'predata_confusionmatrix/2_output_train_gpt4_allint.json'
# file2_name = 'predata_confusionmatrix/B_3_output_train_gpt4_allint_6to1.json'
# target_field = 'label'  # ここにチェックしたいフィールド名を入れる
```

```python
# ファイル名とフィールド名を定義
file1_name = 'predata_test_gpt4/A_test_converted.json'
# file2_name = 'predata_confusionmatrix/2_output_train_gpt4_allint.json'
file2_name = 'predata_test_gpt4/B_2_test_gpt4_allint.json'
target_field = 'label'  # ここにチェックしたいフィールド名を入れる
```

```python

# ファイルを読み込み、データを抽出
data1 = extract_data(load_json(file1_name), target_field)
data2 = extract_data(load_json(file2_name), target_field)

# コンフュージョンマトリックスを計算
cm = confusion_matrix(data1, data2)
print(cm)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

def plot_confusion_matrix(y_preds, y_true, labels):
    cm = confusion_matrix(y_true, y_preds, normalize="true")
    fig, ax = plt.subplots(figsize=(6, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap="Blues", values_format=".2f", ax=ax, colorbar=False)
    plt.title("Normalized confusion matrix")
    plt.show()

#y_preds = lr_clf.predict(X_valid)
labels=["sadness", "joy", "love", "anger", "fear", "surprise"]
plot_confusion_matrix(data2, data1, labels)
```
